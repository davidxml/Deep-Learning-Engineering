### Random Tensors

Why random tensors?

Random tensors are important because the way many neural network learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data.

`Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers.`


```python
# Create a random tensor of size (3, 4)
random_tensor =  torch.rand(3, 4)
random_tensor
```

```python
random_tensor.ndim
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(3, 224, 224)) #colour channels (R, G, B), height, Width
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

```python
# Create a tensor 
some_tensor = torch.rand(3, 4).type(torch.float64)
some_tensor
```

### Matrix Multiplication 

Two main ways of performing  matrix multiplication in  deep learning and deep learning is :
1. Element-wise multiplication
2. Matrix Multiplication (dot product)  

matrix muktiplicaion must satisfy the following conditions:
1. The inner dimensions must match must strictly be  (n by m) * (m by q)
2. The resulting matrix has the shape of the **outer dimensions**
* (n, m) @ (m, n) --> (n, n)
* (3, 2)  @ (2, 9) --> (3, 3)

```python

# Element wise multiplication 
print(tensors, "*", tensors)
print(f" Equals: {tensors * tensors}")
```

```python
# Matrix Multplicaion 
torch.matmul(tensors, torch.tensor([[3], [4], [6]]))
```

```python
%%time
value = 0
for i in range(len(tensors)):
    value += tensors[i] * tensors[i]
print(value)
```

```python
%%time 
torch.matmul(tensors, tensors)
```

```python
### one of the most common errors in deep learning: shape errors 

tensor_a = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_b = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])

torch.mm(tensor_a, tensor_b) # same as torch.matmul
```

### To fix our tensor shape issues, we can manipulate the shape of one of our tensors using the  **transpose**
