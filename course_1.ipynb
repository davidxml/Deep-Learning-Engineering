### Random Tensors

Why random tensors?

Random tensors are important because the way many neural network learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data.

`Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers.`


```python
# Create a random tensor of size (3, 4)
random_tensor =  torch.rand(3, 4)
random_tensor
```

```python
random_tensor.ndim
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(3, 224, 224)) #colour channels (R, G, B), height, Width
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

### Zeros and Ones

### Tensor Datatypes

**Note:** one of the three big errors we'd be running into with Pytorch and deep learning is datatypes 
1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not on the right device 

```python
# Float 32 Tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0], 
                               dtype = torch.float64, # what datatype 
                               device = None,   # what datatype is tensor on
                               requires_grad=False)   # whether or not to track gradients with this tensor operations
                               

float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
float_16_tensor =  float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
torch.cuda.is_available()
```

```python
float_16_tensor * float_32_tensor 
```

```python
int_32_tensor = torch.tensor([3,6,9], dtype=torch.long)
int_32_tensor
```

```python
float_32_tensor * int_32_tensor
```

### Getting information from Tensors (tensor attributes)

1. Tensors not right datatype --> to get datatype from a tensor, can use `tensor.dtype`
2. Tensors not right shape  --> to get shape from a tensor, we can use `tensor.shape`
3. Tensors not on the right device --> to get device from a tensor, can use `tensor.device` 


```python
# Create a tensor 
some_tensor = torch.rand(3, 4).type(torch.float64)
some_tensor
```

```python
# Find out detais about some tensors
print(some_tensor)
print(f"Datatype of tensors: {some_tensor.dtype}")
print(f"Shape of tensor: {some_tensor.shape}")
print(f"Device tensor is on: {some_tensor.device}")
```
